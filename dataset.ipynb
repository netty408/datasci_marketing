### Normalizing tranforming features with MINMAXScalar() and fit_tranform()

```python
cars = pd.read_csv("/Users/netty408lp/Desktop/datasci_cu/mtcars.csv")
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','grar','crab']
cars.head()
```

```python
mpg =cars.mpg
plt.plot(mpg)
```

```python
cars[['mpg']].describe()
```

```python
mpg_matrix = mpg.values.reshape(-1,1)
scaled = preprocessing.MinMaxScaler()
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
standardized_mpg = scale(mpg,axis=0,with_mean=False,with_std=False)
plt.plot(standardized_mpg)
```

```python
standardized_mpg = scale(mpg)
plt.plot(standardized_mpg)
```
